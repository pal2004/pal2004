#Opening a csv file from URL

```python
import requests
from io import StringIO
url=""
headers={"User-Agent": "Mozilla/5.0 (Macintosh; Intel Mac OS X 10.14; rv:66.0) Gecko/20100101 Firefox/66.0"}
req=requests.get(url, headers=headers)
data= StringIO(req.text)
pd.read_csv(data)
```

#Sep Parameter

```python
pd.read_csv('movie_titles_metadata.tsv',sep='\t')
```

```python
#In case you do not have a column names, and you first row acts as column names
pd.read_csv('movie_titles_metadata.tsv',sep='\t',names=['sno','name','release_year','ratings','votes','genres'])
```

# Index_col parameter

```python
#any column that is of no use and want to be converted as index
pd.read_csv('/content/data.csv',index_col='Duration')
```

#Header Parameter

```python
#first row column ka naam hai, but since the data generated by the user marked the column as 0 so pandas can not understand that it is a headers
pd.read_csv('/content/data.csv',header=1)
```

#usecols parameter

```python
#only keeps necessary columns
pd.read_csv('/content/data.csv',usecols=['Duration','Pulse'])
```

#Squeeze Parameter
####squeeze paramter has been removed form read_csv

```python
df=pd.read_csv('/content/data.csv')
df=df.squeeze('columns')
print(df)
```

#Skiprows/nrows Parameter

```python
#used to skip rows
pd.read_csv('/content/data.csv',skiprows=[0,5])
```

```python
#only gets the n number of rows
pd.read_csv('/content/data.csv',nrows=100)
```

#Encoding Parameter

```python
pd.read('zomato_csv',encoding='latin-1')
```

#Skip bad lines
##Expected 5 fields in this line, got 8

```python
pd.read('zomato_csv',encoding='latin-1',error_bad_lines=False)
```

#dtypes Parameter
#converts one data type to another
pd.read_csv('',dtype={'target':int})

#Handling Dates
pd.read('zomato_csv',encoding='latin-1', parse_dates=['date'])
